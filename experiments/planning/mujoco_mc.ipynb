```python
import gym
import numpy as np
import pandas as pd
import time

from aprl.agents.monte_carlo import MujocoResettableWrapper, MonteCarloParallel, receding_horizon
from experiments.common import set_seeds, make_env, fit_ilqr, \
                               on_iteration, receding, multi_evaluate
```

```python
def make_env(env_name, seed):
    env = gym.make(env_name)
    env = MujocoResettableWrapper(env.unwrapped)
    env.seed(seed)
    return env
```

```python
env_name = 'HalfCheetah-v2'
seed = 42
```

```python
env = make_env(env_name, seed=seed)
env_fns = [lambda: make_env(env_name, seed=seed) for _ in range(24)]
mc = MonteCarloParallel(env_fns, 10, 4096, seed=seed)
```

```python
rews = []
for a, ob, rew, done, info in receding_horizon(mc, env):
    print(a, ob, rew, done, info)
    rews.append(rew)
    env.render()
    if len(rews) == 100:
        break
print(sum(rews))
```

```python
mc.close()
```
